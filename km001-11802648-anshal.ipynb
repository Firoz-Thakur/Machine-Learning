## Linear Regression 

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
#training
lr.fit(data_train,tar_train)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
pred_train_lr=lr.predict(data_train)
print(mean_squared_error(pred_train_lr,tar_train))
print(r2_score(pred_train_lr,tar_train))
```

## MEan one

```python
#testing
```

```python
pred_test_lr=lr.predict(data_test)
```

```python
print(mean_squared_error(pred_test_lr,tar_test))
print(r2_score(pred_test_lr,tar_test))
```

## Poly

```python
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures()
```

```python
p_train_d=poly.fit_transform(data_train)
p_test_d=poly.fit_transform(data_test)
```

```python
#Train
pl=LinearRegression()
pl.fit(p_train_d,tar_train)
```

```python
pred_train_pl=pl.predict(p_train_d)
```

```python
print(mean_squared_error(pred_train_pl,tar_train))
print(r2_score(pred_train_pl,tar_train))
```

```python
#Testing
pred_test_pl=pl.predict(p_test_d)
```

```python
print(mean_squared_error(pred_test_pl,tar_test))
print(r2_score(pred_test_pl,tar_test))
```
