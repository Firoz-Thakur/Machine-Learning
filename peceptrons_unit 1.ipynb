```python
import numpy as np
```

```python
x1=[0,0,1,1]
x2=[0,1,0,1]
target=[0,0,0,1]
threshold=0
lr=1
w0=w1=w2=1
epochs=10
x0=1
```

```python
for i in range(epochs):
        print("Iteration:",i+1)
        for j in range(len(x1)):
            s=0
            print("input",x1[j],x2[j])
            s=w0+x1[i]*w1+x2[i]*w2
            if s>=threshold:
                output=1
            else:
                output=0
            error=target[j]-output
            if error!=0:
                w0=w0+lr*x0
                w1=w1+lr*x1[j]
                w2=w2+lr*x2[j]
            print("updated weights",w0,w1,w2)    
            
```
