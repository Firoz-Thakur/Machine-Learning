```python
import numpy as np
x1=[0,0,1,1]
x2=[0,1,0,1]
target=[0,0,0,1]
w0=w1=w2=1
x0=1
epoch=10
lr=1
th=0
```

```python
for i in range(epoch):
    print("iteration",i)
    for j in range(len(x1)):
        s=0
        print("inputs",x1[j],x2[j])
        s=w0*x0 +x1[j]*w1+x2[j]*w2
        if(s>=th):
            out=1
        else:
            out=0 
        print(s)    
        error=target[j]-out
        print("erorr",error)
        if(error!=0):
            w0=w0+lr*error*x0
            w1=w1+lr*error*x1[j]
            w2=w2+lr*error*x2[j]
        print("updated weight",w0,w1,w2)
```
